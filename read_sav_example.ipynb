# Read sav file and extract metadata

```python
import pyreadstat
import numpy as np
data_path = "/User/olaf"
f = "SHEDSname.sav"
df, metadata = pyreadstat.read_sav(f"{data_path}/{f}", encoding="UTF-8")
```

```python

print("\nColumn names:", metadata.column_names)

metadata.variable_value_labels['accom11_1']
```

```python
print("\nColumn labels:", metadata.column_labels)
```

```python
print("\nValue labels:", metadata.variable_value_labels)
```

```python

print("\nVariable types:", metadata.readstat_variable_types)
```

```python
# Filter out un-interesting columns
# Select elements between start and end in the column names inclusive
# Find the index of the start and end elements
lst = df.columns.tolist()
print(f"Columns {lst}")
start_idx = lst.index('start')
end_idx = lst.index('end')
cols_to_drop = lst[start_idx:end_idx + 1]
print(f"Columns to drop: {cols_to_drop}")
```

```python
df = df.drop(columns=cols_to_drop)
df.columns
```

```python
# Create type dict of types and replace type
# Replace double: float64, string: str
replace_types = {'double': np.float64, 'string': str }
types = {k: replace_types.get(v, v) for k, v in metadata.readstat_variable_types.items()}
# Change types according to the variables types from SPSS
for key, value in types.items():
    print(f"for {key} as {types[key]}")
    if key in df.columns.tolist():
        df[key].astype(types[key])
```
